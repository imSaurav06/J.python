```python
# ==========================================================

# What are Lists in Python?
# -------------------------
# A list is a built-in Python data structure used to store
# a collection of items (elements) in a specific order.
#
# Features:
# - Can store multiple values in a single variable
# - Items are ordered
# - Items can be changed (mutable)
# - Can contain different data types
# - Can even contain other lists

# Creating a List
my_list = [1, 2, 3]

# Example:
fruits = ["apple", "banana", "mango"]


# ==========================================================
# Properties of Lists in Python
# ==========================================================

# 1. Ordered
# ----------
# Lists keep items in the order they are added.

my_list = [1, 2, 3]

# Output order remains:
# [1, 2, 3]


# 2. Mutable
# ----------
# List items can be modified after creation.
# You can add, remove, or update elements.

numbers = [10, 20, 30]

# Modify an element
numbers[1] = 25

# Result:
# [10, 25, 30]


# 3. Heterogeneous
# ----------------
# A list can store different data types together.

mixed_list = [1, "hello", 3.14, [5, 6]]

# Elements:
# 1         -> Integer
# "hello"   -> String
# 3.14      -> Float
# [5, 6]    -> List


# 4. Indexed
# ----------
# Every item in a list has an index position.
# Indexing starts from 0.

colors = ["red", "green", "blue"]

# Index positions:
# red    -> 0
# green  -> 1
# blue   -> 2

# Accessing elements
first_color = colors[0]   # red
second_color = colors[1]  # green


# ==========================================================
# Usage of Lists in Data Science
# ==========================================================

# 1. Storing Data Points
# ----------------------
# Lists can store datasets or collections of values.

# Example: Temperatures recorded over a week
temperatures = [22.4, 21.8, 23.0, 24.5, 22.0, 20.5, 19.8]

# Example: Feature names in a dataset
features = ["age", "height", "weight", "income"]


# ==========================================================
# Summary
# ==========================================================

# List:
# - Stores multiple items in one variable
# - Ordered
# - Mutable
# - Heterogeneous
# - Indexed
# - Widely used in Data Science for storing data collections
```
