````python
```python
# ===========================================
# Lecture 17 - Tuples in Python (Part 1)
# ===========================================

# --------------------------------------------
# What are Tuples?
# --------------------------------------------

# Tuples are used to store multiple items in a single variable.
#
# Tuple is one of the 4 built-in collection data types in Python:
# 1. List
# 2. Tuple
# 3. Set
# 4. Dictionary 
#
# A tuple is:
# - Ordered
# - Unchangeable (Immutable)
#
# Tuples are similar to lists, but once created,
# their elements cannot be changed, added, or removed.
#
# Example:
tuple_data = (10, 20, 30)

# --------------------------------------------
# Properties of Tuples
# --------------------------------------------

# 1. Ordered
# Tuples maintain the order of elements.
# Elements can be accessed using indexes.

# Example:
 t = ("apple", "banana", "mango")
 print(t[0])   # apple

# --------------------------------------------

# 2. Immutable
# Once a tuple is created, its values cannot be modified.
#
# Not Allowed:
 t[0] = "orange"
#
# Cannot:
# - Add items
# - Remove items
# - Replace items

# --------------------------------------------

# 3. Heterogeneous Data Types
# A tuple can contain different types of data.

# Example:
 t = (10, 3.14, "hello", True)

# --------------------------------------------

# 4. Indexed
# Tuple elements are accessed using zero-based indexing..
#
# First element  -> index 0
# Second element -> index 1
# Third element  -> index 2

# Example:
 t = ("a", "b", "c")
 print(t[1])   # b

# --------------------------------------------

# 5. Iterable
# Tuples can be traversed using loops.

# Example:
 t = (10, 20, 30)

 for item in t:
     print(item)

# --------------------------------------------

# 6. Fixed Size
# The size of a tuple is fixed after creation.
#
# You cannot:
# - append()
# - remove()
# - pop()
#
# Example:
 t = (1, 2, 3)
 Length remains fixed.

# -------------------------------------------
# Summary
# -------------------------------------------

# Tuple Properties:
#
# ✓ Ordered
# ✓ Indexed
# ✓ Iterable
# ✓ Fixed Size
# ✓ Heterogeneous
# ✓ Immutable
#
# Syntax:
  tuple_name = (item1, item2, item3)

# Example:
  students = ("Ram", "Shyam", "Mohan")
```


````

Creating Tuples

````python
a = (2,3,4,5)

print(a)
print(type(a))
````

````python
tup = ()
tup1 = (5)

print(tup1)
print(type(tup1))
print(tup)
print(type(tup))
````

````python
b = (5,)

print(b)
print(type(b))
````

````python
# a tuple with mixed data types
b = (1, 2.3, "swati", "nishant", True)

print(b)
print(type(b))
````

nested tuples

````python
a = ("mouse", 2.3, (10,20,30))

print(a)
print(type(a))
````

````python
num = ([12,14,16], [20,30], [14,78,96])

print(num)
print(type(num))
````

Tuple Constructor

````python
list = [3,4,5,6]
tuple = tuple(list)

print(tuple)
print(type(tuple))
````

Converting string into a tuple

````python
string_example = "hello"

print(string_example)
print(type(string_example))

tuple11 = tuple(string_example)

print(tuple11)
print(type(tuple11))
````

Converting dictionary into a tuple

````python
dict_example = {
    'a':1, 
    'b':2, 
    'c':3
    }

print(dict_example)
print(type(dict_example))

tup = tuple(dict_example)

print(tup)
print(type(tup))
````

Converting a set into a tuple

````python
set_example = {1,2,3,4}

print(set_example)
print(type(set_example))

tup2 = tuple(set_example)

print(tup2)
print(type(tup2))
````
