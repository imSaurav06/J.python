Practice Question 2 – Data Analysis
Problem Statement

Calculate average score of two classes.

class1_scores = [85,90,88,92,78]
class2_scores = [75,82,80,85,79]

```python
class1_scores = [85,90,88,92,78]
class2_scores = [75,82,80,85,79]

class1_avg = sum(class1_scores)/len(class1_scores)

class2_avg = sum(class2_scores)/len(class2_scores)

print("the avg score of class 1 is", class1_avg)
print("the avg score of class 2 is", class2_avg)
```
