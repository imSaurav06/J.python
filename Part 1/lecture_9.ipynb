# Lecture 9 – For Loop in Python

```python
#Example 1: Iterate Through a List
numbers = [1,2,3,4,5]

for i in numbers:
    print(i)
```

```python
#Example 1: Iterate Through a string
name = "John"

for i in name:
    print(i)
```

```python
# xample 3: range()
for i in range(11):
    print(i)
```

```python
#Example 5: range() with start and end-1
for x in range(5,26):
    print(x)
```

```python
#For Loop with If-Else
for i in range(0,26):

    if i % 2 == 0:
        print("the number is even", i)

    else:
        print("the number is odd", i)
```

```python
#Problem Statement

#Implement a Python program to generate the multiplication table of a given number using a for loop.

number = int(input("enter a number to generate its multiplication table"))

print("The multiplication table of", number)

for i in range(1,11):
    print(f"{number} x {i} = {number*i}")
```

```python
#Break Statement

#Used to terminate a loop when a condition becomes true.

for i in range(5):

    if i == 3:
        break

    print(i)
```

```python
#Continue Statement

#Used to skip the current iteration and move to the next iteration.

for i in range(5):

    if i == 3:
        continue
    
    print(i)

print("i am learning python")
```
