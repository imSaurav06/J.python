# Practice Question

Exam Score Analysis

Write a user-defined function to process exam scores and calculate:

Average exam score
Highest exam score
Lowest exam score
Number of students who passed the exam
Number of students who failed the exam

Passing score = 80 or above

```python


def Score_Analysis(score):
    Number_of_Students= len(score)
    Highest_Score=  max(score)
    Lowest_Score=min(score)
    Average_Score= sum(score)/len(score)
    Number_of_Students_Passed= 0
    # Number_of_Students_Passed= len([i for i in score if i >= 50])
    for i in score:
        if i >= 50:
            Number_of_Students_Passed += 1
    number_of_Students_Failed= Number_of_Students - Number_of_Students_Passed

    return Number_of_Students, Highest_Score, Lowest_Score, Average_Score, Number_of_Students_Passed, number_of_Students_Failed        

```

```python
score = [20, 70, 75, 50, 60, 70, 80, 90, 100]

Number_of_Students, Highest_Score, Lowest_Score, Average_Score, Number_of_Students_Passed, number_of_Students_Failed = Score_Analysis(score)

print("the number of students is", Number_of_Students)
print("the highest score is", Highest_Score)
print("the lowest score is", Lowest_Score)
print("the average score is", Average_Score)
print("the number of students passed is", Number_of_Students_Passed)
print("the number of students failed is", number_of_Students_Failed )
```
