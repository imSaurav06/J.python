````python
# ```python
# ==========================================
# PYTHON BUILT-IN FUNCTIONS CHEAT SHEET
# (Data Science & Machine Learning)
# ==========================================

# int()       -> Convert value to Integer
# float()     -> Convert value to Float
# str()       -> Convert value to String
# bool()      -> Convert value to Boolean

# Examples:
# int("10")      = 10
# float("10.5")  = 10.5
# str(100)       = "100"
# bool(1)        = True


# ==========================================
# DATA TYPE & INFORMATION FUNCTIONS
# ==========================================

# type()      -> Returns data type
# len()       -> Returns number of elements
# id()        -> Returns memory address

# Examples:
# type(10)
# len([1,2,3,4])
# id(x)


# ==========================================
# MATHEMATICAL FUNCTIONS
# ==========================================

# abs()       -> Absolute value
# round()     -> Round number
# pow()       -> Power calculation
# sum()       -> Sum of values
# min()       -> Smallest value
# max()       -> Largest value
# divmod()    -> Quotient & remainder

# Examples:
# abs(-15)
# round(3.14159,2)
# pow(2,3)
# sum([10,20,30])
# min([1,2,3])
# max([1,2,3])
# divmod(10,3)


# ==========================================
# SEQUENCE FUNCTIONS
# ==========================================

# range()      -> Generate sequence
# enumerate()  -> Index + Value
# zip()        -> Combine iterables

# Examples:
# range(5)
# enumerate(['a','b','c'])
# zip(names,scores)


# ==========================================
# COLLECTION FUNCTIONS
# ==========================================

# list()      -> Create List
# tuple()     -> Create Tuple
# set()       -> Unique values only
# dict()      -> Create Dictionary

# Examples:
# list("ABC")
# tuple([1,2,3])
# set([1,1,2,3])
# dict(name="John")


# ==========================================
# SORTING FUNCTIONS
# ==========================================

# sorted()    -> Sort values
# reversed()  -> Reverse values

# Examples:
# sorted([5,2,8,1])
# reversed([1,2,3])


# ==========================================
# LOGICAL FUNCTIONS
# ==========================================

# all()       -> True if all values True
# any()       -> True if any value True

# Examples:
# all([True,True,True])
# any([False,False,True])


# ==========================================
# CHARACTER FUNCTIONS
# ==========================================

# ord()       -> Character to ASCII
# chr()       -> ASCII to Character

# Examples:
# ord('A')
# chr(65)


# ==========================================
# INPUT / OUTPUT FUNCTIONS
# ==========================================

# input()     -> Take user input
# print()     -> Display output

# Examples:
# name = input("Enter Name:")
# print(name)


# ==========================================
# MOST USED IN DATA SCIENCE & ML
# ==========================================

# int()
# float()
# str()
# bool()
# type()
# len()
# abs()
# round()
# sum()
# min()
# max()
# range()
# enumerate()
# zip()
# list()
# tuple()
# set()
# dict()
# sorted()
# all()
# any()

# ==========================================
# END OF NOTES
# ==========================================

````

# ```abs function example

````python

print(abs(-5))
print(abs(5))
print(abs(3+4j)) # because sqrt(3^2 + 4^2) = 5 mean
print(abs(-3-4j)) # because sqrt((-3)^2 + (-4)^2) = 5 mean
print(abs(-22.8654))
````
