# ==========================================
# sum() FUNCTION
# ==========================================

# Definition:
# The sum() function is used to calculate the
# sum of all elements in an iterable such as
# a list, tuple, or set.

# Syntax:
# sum(iterable, start)


```python
my_list = [7,8,9,5,8,629]
result = sum(my_list)
print(result)
```

```python
# Example 2: Using Starting Value

a = [1,2,3,4,5]
total = sum(a, 10) # 10 is the starting value, so it will add 10 to the sum of the list
print(total)
```

```python

# To join strings use +

a = "swati"
b = "nishant"

print(a + b)
```

# ==========================================
# eval() FUNCTION
# ==========================================

# Definition:
# eval() evaluates and executes a Python
# expression stored as a string.

# Syntax:
# eval(expression)

# expression -> String containing a valid
# Python expression

```python
# ==========================================
# Example 2
# ==========================================

exp = "5/2 + 3.5"

print(eval(exp))
print(type(exp))
```

```python
exp = "234*25 + 63 - 98/2"
print(type(exp))
print(eval(exp))
x = int(eval(exp))
print(type(x))
```
