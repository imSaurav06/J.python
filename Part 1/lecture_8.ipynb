# Program: Simulate Exponential Population Growth

# Problem Statement

Develop a Python program to simulate exponential growth.

Prompt the user to enter:

Initial population size
Growth rate
Threshold population size

Use a while loop to simulate the growth until the population reaches a certain threshold.

```python
initial_population = int(input("Enter the initial population: "))
growth_rate = float(input("Enter the growth rate (as a decimal): "))
threshold_population = int(input("Enter the threshold population: ")) #target population
i = 0

population = initial_population

while population < threshold_population:
    population += population * growth_rate
    i += 1

print("The population reached the threshold of this iterations", i)
print(f"Final population size", population)
```

# While Loop using if-else

Problem Statement

You are tasked with validating user input for email addresses.

Write a Python program that prompts the user to enter an email address.

Use a while loop to iterate through the input and check if the email address contains:

"@"
"."

If the email address does not meet these criteria, print:

Invalid email address format.
Please try again.

Otherwise print:

Email address validation passed

```python
email = input("Enter your email address: ")

valis_mail = False

while not valis_mail:
    if "@" in email and "." in email:
        valis_mail = True
        print("Valid email address.")
    else:
        print("Invalid email address. Please try again.")
        email = input("Enter your email address: ")
```
