```python
# ==========================================================
# 3. MULTILINE STRINGS
# ==========================================================

# Triple quotes allow strings to span multiple lines.

text = """hello
world"""

print(text)
```

```python
# ==========================================================
# 4. UPPER METHOD
# ==========================================================

# upper() converts all letters to uppercase.

s = "this is the iscale"
print(s.upper())

# Output:
# THIS IS THE ISCALE
```

```python
# ==========================================================
# 5. LOWER METHOD
# ==========================================================

# lower() converts all letters to lowercase.

s = "THIS IS MY PYTHON CLASS"
print(s.lower())

# Output:
# this is my python class
```

```python
# ==========================================================
# 6. STRING CONCATENATION
# ==========================================================

# Concatenation means joining strings using + operator.

a = "hello"
b = "world"

print(a + " " + b)

# Output:
# hello world
```

```python
# ==========================================================
# 7. STRING REPETITION
# ==========================================================

# * operator repeats a string.

a = "data"
print(a * 5)

# Output:
# datadatadatadatadata


# ==========================================================
# 8. REPLACE METHOD
# ==========================================================

# replace(old, new) replaces occurrences of a substring.

text = "Hello world, I am learning python"

new_text = text.replace("python", "data science")

print(new_text)

# Output:
# Hello world, I am learning data science
```

```python
# ==========================================================
# 9. INDEX METHOD
# ==========================================================

# index() returns the first position of a substring.
# Error occurs if substring is not found.

a = "i want to learn"

print(a.index("to"))

# Output:
# 7
```

```python
# ==========================================================
# 10. FIND METHOD
# ==========================================================

# find() returns the position of a substring.
# Returns -1 if not found.

a = "python is great"

print(a.find("g"))

# Output:
# 10
```

```python
# ==========================================================
# DIFFERENCE BETWEEN index() AND find()
# ==========================================================

# index()
# - Returns position
# - Gives error if not found

# find()
# - Returns position
# - Returns -1 if not found


# ==========================================================
# PRACTICE QUESTION
# ==========================================================

# 1. Concatenate "hello" and "world"
# 2. Find length
# 3. Extract "world"
# 4. Reverse it
# 5. Convert to uppercase
# 6. Replace 'L' with 'X'

concatenated_string = "hello" + "world"

length = len(concatenated_string)

substring = concatenated_string[5:]

reverse = substring[::-1]

upper = reverse.upper()

modify = upper.replace("L", "X")

print(concatenated_string)
print(length)
print(substring)
print(reverse)
print(upper)
print(modify)

# Output:
# helloworld
# 10
# world
# dlrow
# DLROW
# DXROW
```

Quick Revision Table

| Method      | Purpose                                   |
| ----------- | ----------------------------------------- |
| `len()`     | Length of string                          |
| `upper()`   | Convert to uppercase                      |
| `lower()`   | Convert to lowercase                      |
| `replace()` | Replace text                              |
| `index()`   | Position of substring                     |
| `find()`    | Position of substring (`-1` if not found) |
| `+`         | Concatenate strings                       |
| `*`         | Repeat string                             |
| `[::-1]`    | Reverse string                            |
