Slicing with Jump (Step)

```python
a = "i am learning datascience"

print(a[0:10:2])
print(a[0::2])
print(a[::2])
```

Negative Slicing

```python
a = "i am learning datascience"

print(a[-6:])
print(a[::-1])
print(a[-16:])
print(a[-18:-1:1])
print(a[-18:-1:2])
print(a[-18:-1:3])
print(a[2::-1])
```

#Practice Question 1

Write a Python function that takes a date string in the format:

YYYY-MM-DD

```python
def extract_date_parts(date_string):

    year = date_string[0:4]
    month = date_string[5:7]
    day = date_string[8:10]

    return year, month, day
```

```python
extracted_year, extracted_month, extracted_day = extract_date_parts("2024-06-15")
print("Year:", extracted_year)
print("Month:", extracted_month)    
print("Day:", extracted_day)
```

#Practice Question 3

Write a Python function that takes a string as input and returns the substring from the 3rd to the 8th character.

Examples:

Data science
Hello World
All the best

```python
def substring_3_to_8(text):
    return text[3:9]
```

```python
substring = substring_3_to_8("Hello, World!")
print(substring)  # Output: "lo, Wo"
substring = substring_3_to_8("Python Programming")
print(substring)  # Output: "hon Pr"
substring = substring_3_to_8("Data Science is fun!")
print(substring)  # Output: "a Sci"
```
