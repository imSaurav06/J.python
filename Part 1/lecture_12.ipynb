Question 1: Grade Distribution
Problem Statement

Write a function:

grade_distribution(grades)

that takes a list of student grades (0–100) and categorizes them into:

Grade	Range
A	90–100
B	80–89
C	70–79
D	60–69
E	Below 60

Use:

Loop
if-elif-else

Return counts of each category.

```python
def grade_distribution(grades):

    for grade in grades:

        if 90 <= grade <= 100:
            print("you have got A grade")

        elif 80 <= grade <= 89:
            print("you have got B grade")

        elif 70 <= grade <= 79:
            print("you have got C grade")

        elif 60 <= grade <= 69:
            print("you have got D grade")

        else:
            print("you have got E grade")

    return
```

Question 2: Number Guessing Game
Problem Statement

Create a number guessing game.

Computer generates a random number between 1 and 100.
User guesses the number.
Program gives feedback:
Too High
Too Low
Correct
Use:
Functions
Loops
Control Flow Statements

```python
import random

def number_guessing_game():

    secret_number = random.randint(1,100)

   

    print("Welcome to the number guessing game")
    print("I have choosen the number between 1 to 100. Can you guess it ?")

    attempts = 0
    guess = None

    while guess != secret_number:
        guess = int(input("Enter your guess: "))
        attempts += 1

        if guess < secret_number:
            print("Your guess is too low. Try again.")
        elif guess > secret_number:
            print("Your guess is too high. Try again.")
        else:
            print("Congratulations! You guessed the number correctly.")
            print("It took you", attempts, "attempts to guess the number.")

        
```

Question 3: Sum of Squares of First n Natural Numbers
Problem Statement

Create a program to calculate:

1
2
+2
2
+3
2
+...+n
2

Requirements:

User enters n.
Use a loop.
Create a user-defined function.
Call the function from main().

```python
def sum_of_squares(n):

    total = 0

    for i in range(1, n + 1):
        total += i ** 2

    return total


```

```python
def main():

    n = int(input("Enter a positive integer: "))

    result = sum_of_squares(n)

    print(f"The sum of squares of the first {n} natural numbers is: {result}")

main()
```
